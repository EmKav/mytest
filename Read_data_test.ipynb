```python
# This is a test data reading file for the Binder course 31Jan2020


import numpy as np

filename='data.txt'

# Set up list to receive the data
numbers = []

# Read through the file
with open(filename) as openfileobject:
    for line in openfileobject:
        numbers.append(line)
        #print(line)

# Sum the numbers
sum=np.float(0.0)
for i in range(len(numbers)):
    sum=np.float(numbers[i])+sum
    print(sum)
    
    
    
```
